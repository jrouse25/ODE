```python
e = 2.71828182845904523536028747135266249

def euler(x, delta):
    return x + delta * x

def midpoint_method(x, delta):
    k = x + delta / 2 * x
    return x + delta * k

def runge_kutta(x, delta):
    k1 = x
    k2 = x + delta * k1 / 2
    k3 = x + delta * k2 / 2
    k4 = x + delta * k3
    return x + (1/6) * delta * (k1 + 2 * k2 + 2 * k3 + k4)

def execute(f, N):
    x = 1
    for _ in range(0, N):
        x = f(x, 1/N)
    return {"N": N, "x(1)": x, "err": e - x}

gen_table = lambda f: list(map(lambda k: execute(f, 2 ** k), range(10, 21)))

gen_table(euler)
```
