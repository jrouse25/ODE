```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import odeint 

t_range = np.linspace(0, 50, 5000)
```

```python
def vdp(X, t_range, mu):
    dx2 = mu * (1 - X[0]**2)*X[1] - X[0]
    return [X[1], dx2]

def de_solve(X0, mu, t_range):
    res = odeint(vdp, X0, t_range, args=(mu,))
    return res
```

```python
X = [5, 0]
mus = np.arange(0, 5, 0.5)

plt.figure(figsize=(10,6))
for mu in mus:
    
    solution = odeint(vdp, X, t_range, args=(mu,))
    x = solution[:, 0]
    v = solution[:, 1]
    plt.plot(x, v, label=f'mu = {mu}')
    
# Add labels and legend
plt.xlabel('x(t)')
plt.ylabel('v(t)')
plt.legend(loc='upper right')
plt.title('van der Pol solutions in phase space')
plt.show()
```
